```python
import pandas as pd
df = pd.read_json("marc_records.json", )
df
```

```python
# replace empty year with 1800
df["year"] = df["year"].replace("", 1800)
df["year"] = df["year"].replace(5780., 1800)
```

```python
df["year"].unique()
```

```python
df.dropna(subset=["year"], inplace=True)
df["year"] = df["year"].astype(int)
df["year"].hist()
```

```python
lk = df[df["pdf_filename"].str.contains("lk")]
lk
```

```python
ca = df[df["pdf_filename"].str.contains("ca")]
ca
```

```python
books = df[df["pdf_filename"].str.contains("b")]
books
```

```python
# let's check if pdf_filename exists in books folder
not_exist = []
import os
for filename in books["book_id"]:
    if not os.path.exists(f"books/{filename}"):
        print(filename)
        not_exist.append(filename)
len(not_exist)
```

```python
books.dropna(subset=["urls"], inplace=True)
```

```python
# extract the book id, ex. bx000016572 from the list of strings
books["book_id"] = books["urls"].apply(lambda x: x[0].split("/")[-2] + "_0001.pdf")
books
```
